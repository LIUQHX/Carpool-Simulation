```python
# coding: utf-8
import sys, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
```

```python
df = pd.read_csv('./output/res/0109_2000_300_600_windows.csv', index_col=0)
```

```python
df.head()
```

```python
# 
x = df.index.values
xticks = [str(i) for i in x]
```

```python
# Mean Service Rate
y = df.MeanServiceRate.values
plt.close()
fig, ax = plt.subplots(figsize=(8,6), dpi=100, facecolor="white")
ax.plot(x, y, '-o', c='r', lw=2, mew=2, ms=10, mfc='red', mec='r')
ax.tick_params(direction='in', pad=5)
ax.set_ylim([0.66, 0.72])
ax.set_xlabel("Time Interval (s)", {'fontsize': 22, 'fontweight': 'bold'}, labelpad=10)
ax.set_ylabel("Mean Service Rate", {'fontsize': 22, 'fontweight': 'bold'}, labelpad=10)
ax.grid(linestyle=':', color='black')
# ax.set_xticks(x)
# ax.set_xticklabels(xticks, {'fontsize': 15})
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=15)
# plt.show()
fig.savefig("./output/img/windows/MSR.png", bbox_inches='tight')
```
